```python
# 파이썬에서 모듈을 불러올때는 import명령을 사용한다

import urllib.request

url = "https://t1.daumcdn.net/daumtop_chanel/op/20200723055344399.png"
imgName = "C:\yePython\BigData\download\daum.png"

urllib.request.urlretrieve(url, imgName) #urlretrieve (url, 저장할 파일 경로)
print("다운로드 완료")
```

```python
# urlopen 메모리에 먼저 로딩

url = "https://t1.daumcdn.net/daumtop_chanel/op/20200723055344399.png"
imgName = "C:\yePython\BigData\download\daum2.png"

downloadimg = urllib.request.urlopen(url).read()

with open(imgName, "wb") as f: # w는 쓰기모드 , b는 바이너리모드(이미지)
    f.write(downloadimg)

print("다운로드 완료")
```

```python
import urllib.request
import urllib.parse  # url 인코딩

rssUrl = "https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp"

values = {
    'stnId':184
}

params = urllib.parse.urlencode(values)

url = rssUrl + "?" + params

print("url = ",url)

data = urllib.request.urlopen(url).read()
text = data.decode("utf-8")

print(text)
```
