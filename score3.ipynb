```python
# 성적처리 program  # 등록, 검색, 삭제, 수정, 조회, 파일 입출력
import pickle

#과목 설정
subject=['국어','영어','수학','과학']
program = dict()

class Read:
    def read(self,menu,name):
        with open('score2/program.p','rb') as file:
            program=pickle.load(file)
            if menu==5:
                for key, value in program.items():
                    print(key,':', value)
        return program

class Write:
    def write(self,menu,name):
        with open('score2/program.p','wb') as file:
            pickle.dump(program,file)
        return program

class Inputs:
    def inputs(self,menu,name):
        scores = {subject[i]:input(subject[i]+' : ') for i in range(len(subject))}
        program.setdefault(name,scores)

class Select:
    def select(self,menu,name):
        if program.get(name) !=None :
            print(name,'성적')
            print(program.get(name))
        else:
            print('존재하지 않는 학생입니다.')

class Delete:            
    def delete(self,menu,name):
        if program.get(name) !=None :
            program.pop(name)
            if menu==3:
                print('삭제 되었습니다.')
            elif menu==4:
                p=Inputs()
                p.inputs(menu,name)
                print('수정 되었습니다.')
        else:
            print('존재하지 않는 학생입니다.')
            
#####

class Program_input(Read,Inputs,Write): #등록
    pass
class Program_delete(Read,Delete,Write): #수정 / 삭제
    pass
class Program_select(Read,Select): #조회
    pass
    

while True:
    menu=int(input('\n1.등록 | 2.검색 | 3.삭제 | 4.수정 | 5.전체 조회 | 6.종료\n'))

    if menu==1: # 등록
        print('등록하기')
        name=input('이름 : ')
        
        p=Program_input()
        p.read(menu,name)
        p.inputs(menu,name)
        p.write(menu,name)

    if menu==2: # 검색
        print('검색하기')
        name=input('이름 : ')
        
        p=Program_select()
        p.read(menu,name)
        p.select(menu,name)

    if menu==3 or menu==4: # 삭제/수정
        if menu==3:
            print('삭제하기')
        else:
            print('수정하기')
        name=input('이름 : ')
        
        p=Program_delete()
        p.read(menu,name)
        p.delete(menu,name)
        p.write(menu,name)

    if menu==5: # 전체 조회
        print('전체 조회')
        p=Read()
        p.read(menu,name)            
    
    if menu==6: # 종료
        print('종료되었습니다.')
        break
```
