```python
# 성적처리 program  # 등록, 검색, 삭제, 수정, 조회, 파일 입출력
import pickle

#과목 설정
subject=['국어','영어','수학','과학']

def read():
    program = dict()
    with open('score2/program.p','rb') as file:
        program=pickle.load(file)
    return program
    
def write():
    with open('score2/program.p','wb') as file:
        pickle.dump(program,file)
    return program

def inputs(name):
    scores = {subject[i]:input(subject[i]+' : ') for i in range(len(subject))}
    program.setdefault(name,scores)

def select(name):
    if program.get(name) !=None :
        print(name,'성적')
        print(program.get(name))
    else:
        print('존재하지 않는 학생입니다.')
        
def delete(name,menu):
    if program.get(name) !=None :
        program.pop(name)
        if menu==3:
            print('삭제 되었습니다.')
        elif menu==4:
            inputs(name)
            print('수정 되었습니다.')
    else:
        print('존재하지 않는 학생입니다.')


while True:
    menu=int(input('1.등록 | 2.검색 | 3.삭제 | 4.수정 | 5.전체 조회 | 6.종료'))

    if menu==1: # 등록
        print('등록하기')
        name=input('이름 : ')
        read()
        inputs(name)
        write()

    if menu==2: # 검색
        print('검색하기')
        name=input('이름 : ')
        read()
        select(name)

    if menu==3: # 삭제
        print('삭제하기')
        name=input('이름 : ')
        read()
        delete(name,menu)
        write()

    if menu==4: # 수정
        print('수정하기')
        name=input('이름 : ')
        read()
        delete(name,menu)
        write()

    if menu==5: # 전체 조회
        print('전체 조회')
        with open('score2/program.p','rb') as file:
            program=pickle.load(file)

            for key, value in program.items():
                print(key,':', value)
    
    if menu==6: # 종료
        print('종료되었습니다.')
        break
```
