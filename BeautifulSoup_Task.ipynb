```python
# 네이버 뉴스
from bs4 import BeautifulSoup
import requests

url = 'https://news.naver.com/main/list.nhn?mode=LPOD&mid=sec&sid1=001&sid2=140&oid=001&isYeonhapFlash=Y&aid=0011993440'

# 웹주소의 내용을 받아오는 request.get 부분에서 문제가 생기면.. User-Agent를 설정함으로써 해결할 수 있음.
# 참고 : https://backkom-blog.tistory.com/entry/Chrome크롬-웹-브라우저-User-Agent-확인하기
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Whale/2.8.108.15 Safari/537.36'}
html = requests.get(url, headers = headers)

# html 분석하기
soup = BeautifulSoup(html.text, 'html.parser')

# mainTitle
mainTitle = soup.select("dl.type04 > dt")[1]
for a in mainTitle:
    href = a.attrs['href']  # attrs : 속성
    text = a.string
    print(text)
    print("-->" ,href)

print("")    
# subTitle    
subTitle = soup.select("dl.type04 > dd > ul > li > a")
for a in subTitle:
    href = a.attrs['href']  # attrs : 속성
    text = a.string
    print("    ",text)
    print("    -->",href)

print("")
# otherTitle
otherTitle = soup.select("ul.type02 > li > a")
for a in otherTitle:
    href = a.attrs['href']  # attrs : 속성
    text = a.string
    print("    ",text)
    print("    -->",href)
```

```python
# 네이버 증권
from bs4 import BeautifulSoup
import urllib.request as req

url = "https://finance.naver.com/marketindex/"
data = req.urlopen(url).read()

# html 분석하기
soup = BeautifulSoup(data, 'html.parser')

title = soup.select('h3.h_lst')
price = soup.select('span.value')


while True:
    choice = int(input("금융 번호를 입력하세요.(0 ~ 11) : "))
    
    if choice > 11:
        print("종료합니다.")
        break
    
    for i in range(0, 12):
        if i == choice:
            print(title[i].string," : ",price[i].string, "원")
            print()
    
```
